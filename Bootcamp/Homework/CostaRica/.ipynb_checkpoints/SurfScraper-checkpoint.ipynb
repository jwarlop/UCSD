```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
```

```python
# Initialize browser
def init_browser():
    # @NOTE: Replace w/path to chromedriver
    exec_path={"executable_path":"/usr/local/bin/chromedriver"}
    return Browser("chrome",**exec_path,headless=False)
```

```python
browser = init_browser()
```

```python
url = "https://www.surfline.com/surf-reports-forecasts-cams/costa-rica/3624060"
browser.visit(url)
```

```python
# Scrape page into soup
html = browser.html
soup = bs(html, 'html.parser')

# Find today's forecast
#forecast_today = soup.find("div", class_="weather-forecasts todays-weather forecast")
#forecast_today
```
