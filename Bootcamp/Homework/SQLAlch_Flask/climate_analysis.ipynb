```python
def calc_temps(sdate,edate,conn):
    stmt = "SELECT `tobs` FROM HMEASUREMENTS WHERE date "
    stmt += "BETWEEN '{}' AND '{}'".format(sdate,edate)
    rs = conn.execute(stmt)
    l = rs.fetchall()
    x = [i[0] for i in l]
    avg,min_,max_ = sum(x)/len(x),min(x),max(x)
    return( (avg,min_,max_))
```
